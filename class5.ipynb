```python
#WAP TO CREATE A CALCULATON USING FUNCTION
def calculator():
    x=int(input("enter  number x="))
    y=int(input("enter number y="))
    o=input("enter an operator")
    if o=="+":
        return x+y
    elif o=="-":
        return x-y
    elif o=="*":
        return x*y
    elif o=="/" and y!=0:
        return x/y
    elif o=="/" and y==0:
        print("the value of y cannot be zero")
    else:
        print("invalid number")
calculator()
```

```python
def add(x,y):
    return x+y
def sub(x,y):
    return x-y
def mul(x,y):
    return x*y
def div(x,y):
    return a/b

def calculator(x,y,o):
    
    if o=="+":
        print(add(x,y))
    elif o=="-":
        print(sub(x,y))
    elif o=="*":
        print(mul(x,y))
    elif o=="/" and y!=0:
        print(div(x,y))
    elif o=="/" and y==0:
        print("the value of y cannot be zero")
    else:
        print("invalid number")
x=int(input("enter  number x="))
y=int(input("enter number y="))
o=input("enter an operator")
calculator(x,y,o)
```
